##**9. Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Dari Persentase Ouliers diatas ada beberapa kolom yang memiliki ouliers yaitu kolom berat_sampah_kg memiliki persentase terbesar (17,4%), Frekuensi_Pengangkutan_per_Minggu (11,6%), dan Daur_Ulang_Kg 	(3,8%). Hanya variabel Luas_Area_TPS_m2 dan jumlah_penduduk memiliki data yang sangat stabil dan konsisten (0%).

#**Eksplorasi Data (EDA)**

##**1. Comparison(Perbandingan)**

###**a. Perbandingan Total Berat Sampah per Provinsi**

```python
berat_provinsi = df.groupby('Provinsi')['berat_sampah_kg'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=berat_provinsi.index, y=berat_provinsi.values)
plt.xticks(rotation=45)
plt.title('Total Berat Sampah per Provinsi')
plt.ylabel('Total Berat Sampah (kg)')
plt.show()
```

Berdasarkan grafik tersebut, terlihat bahwa beberapa nama provinsi muncul lebih dari satu kali, yang menunjukkan adanya ketidakkonsistenan penulisan data (misalnya perbedaan huruf besarâ€“kecil atau adanya spasi tersembunyi). Hal ini menyebabkan total berat sampah untuk provinsi yang sama terpecah menjadi beberapa kategori, sehingga hasil perbandingan menjadi kurang akurat. Sebagai contoh, jika â€œJawa Timurâ€ ditulis dengan format berbeda, maka jumlah sampahnya tidak akan terakumulasi secara maksimal dalam satu batang grafik. Kondisi ini menunjukkan bahwa sebelum melakukan analisis lebih lanjut, perlu dilakukan pembersihan data agar setiap provinsi memiliki satu format penulisan yang konsisten. Dengan data yang sudah dibersihkan, perbandingan total berat sampah antar provinsi akan lebih valid dan dapat digunakan sebagai dasar pengambilan keputusan yang lebih tepat.

##**2. Composition(Komposisi)**

```python
komposisi = df.groupby('Jenis_Sampah')['berat_sampah_kg'].sum()
komposisi_persen = (komposisi / komposisi.sum()) * 100

plt.figure(figsize=(8,8))
komposisi_persen.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Komposisi Berat Sampah Berdasarkan Jenis (%)')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

Berdasarkan grafik komposisi tersebut, terlihat bahwa kategori B3 (19,8%) dan Campuran (18,1%) menjadi penyumbang terbesar terhadap total berat sampah. Selanjutnya terdapat kategori Anorganik (10,1%), diikuti beberapa variasi penulisan lain seperti Organik (8,0%), ORGANIK (7,6%), dan organik (5,5%), serta variasi anorganik dan ANORGANIK dengan persentase sekitar 6â€“6,6%. Selain itu, terdapat kategori campuran (8,2%) dan CAMPURAN (4,4%), serta b3 (5,3%).

Namun, grafik ini menunjukkan adanya duplikasi kategori akibat perbedaan huruf besarâ€“kecil. Jika kategori dengan nama yang sama digabungkan (misalnya seluruh â€œOrganikâ€ menjadi satu, seluruh â€œCampuranâ€ menjadi satu, dan seluruh â€œB3â€ menjadi satu), maka proporsi sebenarnya kemungkinan jauh lebih besar pada kategori tertentu. Artinya, secara umum komposisi sampah didominasi oleh B3 dan Campuran, tetapi hasil ini masih perlu pembersihan data agar benar-benar merepresentasikan kondisi sebenarnya.

##**4. Relationship/Hubungan**

```python
kolom_analisis = ['berat_sampah_kg', 'Luas_Area_TPS_m2', 'Daur_Ulang_Kg', 'Frekuensi_Pengangkutan_per_Minggu']

plt.figure(figsize=(10, 8))
sns.heatmap(data=df[kolom_analisis].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')

plt.title('Korelasi antar Variabel Pengelolaan Sampah')
plt.show()
```

Insight utama dari heatmap korelasi ini adalah variabel-variabel pengelolaan sampah bersifat independen atau tidak saling memengaruhi, di mana angka korelasi yang mendekati nol menunjukkan bahwa kenaikan volume sampah tidak otomatis diikuti dengan peningkatan jumlah daur ulang maupun penyesuaian frekuensi pengangkutan. Hal ini mengindikasikan bahwa sistem operasional di lapangan kemungkinan besar masih berjalan berdasarkan jadwal rutin yang kaku dibandingkan sistem yang responsif terhadap data volume nyata, sehingga diperlukan evaluasi lebih lanjut untuk menyelaraskan kapasitas TPS dengan beban sampah yang sebenarnya dihasilkan.
