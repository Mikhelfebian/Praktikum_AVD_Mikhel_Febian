## **Pengecekan Outliers**

```python
results = []

cols = baca_dataset.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = baca_dataset[col].quantile(0.25)
  q3 = baca_dataset[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = baca_dataset[(baca_dataset[col] < lower_bound) | (baca_dataset[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(baca_dataset))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Dari pengecekan outliers diatas ada beberapa kolom yang memiliki outliers yaitu:
1. math score	sebanyak 0.8%
2. reading score	sebanyak 0.6%
3. writing score	sebanyak 0.5%

# **Eksplorasi Data (EDA)**

## **Comparison/Perbandingan**

```python
baca_dataset['average_score'] = (
    baca_dataset['math score'] +
    baca_dataset['reading score'] +
    baca_dataset['writing score']
) / 3

prep_avg = baca_dataset.groupby('test preparation course')['average_score'].mean()

plt.figure(figsize=(6,4))
sns.barplot(x=prep_avg.index, y=prep_avg.values)
plt.title('Pengaruh Test Preparation Course terhadap Nilai')
plt.xlabel('Test Preparation Course')
plt.ylabel('Rata-rata Nilai')
plt.show()
```

perbandingan antara Test Preparation Course dan nilai rata-rata

- Siswa yang menyelesaikan *test preparation course* memiliki rata-rata nilai lebih tinggi dibandingkan yang tidak mengikuti.  
- Terdapat selisih nilai yang menunjukkan dampak positif program persiapan terhadap hasil ujian.

Test preparation course berkontribusi terhadap peningkatan performa akademik dan dapat menjadi strategi yang efektif.


```python
gender_avg = baca_dataset.groupby('gender')['average_score'].mean()

plt.figure(figsize=(6,4))
sns.barplot(x=gender_avg.index, y=gender_avg.values)
plt.title('Rata-rata Nilai Berdasarkan Gender')
plt.xlabel('Gender')
plt.ylabel('Rata-rata Nilai')
plt.show()
```

Perbandingan antara gender dan nilai rata-rata

- Rata-rata nilai antara laki-laki dan perempuan relatif tidak berbeda secara ekstrem.
- Siswa perempuan cenderung lebih unggul pada reading dan writing.  
- Siswa laki-laki cenderung sedikit lebih tinggi pada matematika.

Gender bukan faktor dominan dalam menentukan prestasi akademik secara keseluruhan, namun terdapat kecenderungan performa berbeda pada jenis mata pelajaran tertentu.


```python
parent_avg = baca_dataset.groupby('parental level of education')['average_score'].mean().sort_values()

plt.figure(figsize=(10,5))
sns.barplot(x=parent_avg.index, y=parent_avg.values)
plt.title('Pengaruh Pendidikan Orang Tua terhadap Nilai')
plt.xlabel('Pendidikan Orang Tua')
plt.ylabel('Rata-rata Nilai')
plt.xticks(rotation=45)
plt.show()
```

Perbandingan antara Parental Level of Education dan nilai rata-rata

- Semakin tinggi tingkat pendidikan orang tua, semakin tinggi pula rata-rata nilai siswa.  
- Siswa dengan orang tua berpendidikan sarjana atau master menunjukkan performa akademik yang lebih baik dibandingkan kelompok lainnya.

Latar belakang pendidikan orang tua memiliki hubungan dengan prestasi akademik siswa.

```python
lunch_avg = baca_dataset.groupby('lunch')['average_score'].mean()

plt.figure(figsize=(6,4))
sns.barplot(x=lunch_avg.index, y=lunch_avg.values)
plt.title('Perbandingan Nilai Berdasarkan Jenis Lunch')
plt.xlabel('Jenis Lunch')
plt.ylabel('Rata-rata Nilai')
plt.show()
```

Lunch vs Average Score

- Siswa dengan kategori *standard lunch* memiliki rata-rata nilai lebih tinggi dibandingkan siswa dengan *free/reduced lunch*.  
- Perbedaan ini mengindikasikan kemungkinan adanya pengaruh faktor sosial-ekonomi terhadap performa akademik.

Terdapat indikasi kesenjangan performa berdasarkan latar belakang ekonomi siswa.

## **Relationship/Hubungan**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    baca_dataset[['math score', 'reading score', 'writing score']].corr(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Korelasi Antar Nilai Siswa')
plt.show()
```

Reading dan Writing biasanya memiliki korelasi paling tinggi, karena keduanya termasuk kemampuan literasi yang saling berkaitan.

Korelasi positif antar semua mata pelajaran menunjukkan bahwa siswa yang berprestasi di satu bidang cenderung berprestasi di bidang lain.

Hal ini mengindikasikan bahwa performa akademik siswa bersifat konsisten, bukan berdiri sendiri pada satu mata pelajaran saja.
